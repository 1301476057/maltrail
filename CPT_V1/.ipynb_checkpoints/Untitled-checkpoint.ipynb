```python
data = pd.read_csv('lastest_some_deletesame2_Generalization_deletesame1.csv',sep='\a')
```

```python
rand = np.random.randint(7,size=len(data))
```

```python
data.groupby('ip_24').count()['host']
```

```python
data['change'] = rand
```

```python
s = 'family-partners.fr'
s1 = 'jessisjewels.com'
s2 = 'pogruz.wanyizhao.net'
```

```python
def change(x,y,z):
    if(x==0):
#         print(0)
        y = y.replace(z,s)
    elif(x==2):
        y = y.replace(z,s1)
    elif(x==4):
        y = y.replace(z,s2)
    return y
```

```python
data['full_uri'] = data.apply(lambda x:change(x['change'], x['full_uri'], x['original_host']),axis=1)
```

```python
def change1(x,y):
    if(x==0):
#         print(0)
        y = s
    elif(x==2):
        y = s1
    elif(x==4):
        y = s2
    return y
```

```python
data['original_host'] = data.apply(lambda x:change1(x['change'], x['original_host']),axis=1)
```

```python
data[['original_host','full_uri','change']]
```

```python
data.to_csv('lastest_some_deletesame2_Generalization_deletesame1_change.csv')
```
